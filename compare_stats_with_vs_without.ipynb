```python
import pandas as pd
import numpy as np
from dateutil import parser

start_date = None
end_date = None

df = pd.read_csv('large_boxscoretraditionalv2_df.csv', index_col=0)
df.drop_duplicates(inplace=True)
#     print(df.columns)
df.dropna(subset=['MIN'],inplace=True)


df_g = pd.read_csv('all_games.csv')
df_g=df_g[['GAME_ID','TEAM_ID','GAME_DATE']].drop_duplicates().dropna()
df_g['GAME_DATE'] = pd.to_datetime(df_g['GAME_DATE'])
df = df.merge(df_g, 
     how='inner',
     left_on=['GAME_ID','TEAM_ID'],
     right_on=['GAME_ID','TEAM_ID'])


if start_date is not None:
    print('len before filter start date', len(df))
    print (df['GAME_DATE'].max())
    df=df[df['GAME_DATE']>=parser.parse(start_date)]
    print('len after filter  start date', len(df))
if end_date is not None:
    print('len before filter  end date', len(df))
    df=df[df['GAME_DATE']<=end_date]
    print('len after filter  end date', len(df))
    
df['min_round'] = df['MIN'].apply( lambda x: int(x.split(':')[0]))    
df[df['TEAM_CITY']=='Milwaukee']
```

```python



player_name = 'Aaron Holiday'
other_players = ['Jeremy Lamb','Myles Turner']

unique_games_certain_player = df[(df['PLAYER_NAME']==player_name)&
                                (df['GAME_DATE'] >= '2019-10-22')&
                                (df['min_round'] > 10)]['GAME_ID'].unique()

unique_games_other_player = df[(df['PLAYER_NAME'].isin(other_players))&
                                (df['GAME_DATE'] >= '2019-10-22')&
                              (df['min_round'] > 10)]['GAME_ID'].unique()


games_apart    = list(set(unique_games_certain_player) - set(unique_games_other_player))
games_together = list(set(unique_games_certain_player).intersection(unique_games_other_player))
number_of_games_apart = len(games_apart)
number_of_games_together = len(games_together)

games_apart_df = df[(df['PLAYER_NAME']==player_name)&
                    (df['GAME_DATE'] >= '2019-10-22')&
                      df['GAME_ID'].isin(games_apart)]

games_together_df = df[(df['PLAYER_NAME']==player_name)&
                    (df['GAME_DATE'] >= '2019-10-22')&
                      df['GAME_ID'].isin(games_together)]

games_apart_df_desc    = games_apart_df.describe()[['FGM','FGA','FTM','FTA','FG3M','PTS','REB','AST','STL','BLK','TO','PLUS_MINUS']]
games_together_df_desc = games_together_df.describe()[['FGM','FGA','FTM','FTA','FG3M','PTS','REB','AST','STL','BLK','TO','PLUS_MINUS']]

(games_apart_df_desc.ix['mean'] - games_together_df_desc.ix['mean']).to_frame().round(3)
```

```python
import itertools
list_of_teams = df['TEAM_CITY'].unique() 
large_list_of_results = []

for team in list_of_teams:
    print(team)
    list_players = df[df['TEAM_CITY']==team]['PLAYER_NAME'].unique()
    list_of_pairs = list(itertools.combinations(list_players, 2))


    for pair_of_players in list_of_pairs:
        unique_games_certain_player = df[(df['PLAYER_NAME']==pair_of_players[0])&
                                        (df['GAME_DATE'] >= '2019-10-22')&
                                         (df['min_round'] > 11)]['GAME_ID'].unique()

        unique_games_other_player = df[(df['PLAYER_NAME']== pair_of_players[1])&
                                        (df['GAME_DATE'] >= '2019-10-22')&
                                       (df['min_round'] > 11)]['GAME_ID'].unique()


        games_apart    = list(set(unique_games_certain_player) - set(unique_games_other_player))
        games_together = list(set(unique_games_certain_player).intersection(unique_games_other_player))
        number_of_games_apart = len(games_apart)
        number_of_games_together = len(games_together)

        if (number_of_games_apart > 6 )&(number_of_games_together>6):
            games_apart_df = df[(df['PLAYER_NAME']==pair_of_players[0])&
                        (df['GAME_DATE'] >= '2019-10-22')&
                          df['GAME_ID'].isin(games_apart)]

            games_together_df = df[(df['PLAYER_NAME']==pair_of_players[0])&
                        (df['GAME_DATE'] >= '2019-10-22')&
                          df['GAME_ID'].isin(games_together)]
            games_apart_df_desc    = games_apart_df.describe()[['FGM','FGA','FTM','FTA','FG3M','PTS','REB','AST','STL','BLK','TO','PLUS_MINUS']]
            games_together_df_desc = games_together_df.describe()[['FGM','FGA','FTM','FTA','FG3M','PTS','REB','AST','STL','BLK','TO','PLUS_MINUS']]

            pts_mean = (games_apart_df_desc.ix['mean'] - games_together_df_desc.ix['mean']).to_frame().round(3).ix['PTS'].values[0]
            plus_minus_mean = (games_apart_df_desc.ix['mean'] - games_together_df_desc.ix['mean']).to_frame().round(3).ix['PLUS_MINUS'].values[0]
            row = [pair_of_players[0], pair_of_players[1] , number_of_games_apart, number_of_games_together , pts_mean , plus_minus_mean]        
            large_list_of_results.append(row)
```

```python
df_compare = pd.DataFrame(large_list_of_results , 
                          columns = ['curr_player' , 'opp_player', 'games_apart', 'games_together', 'mean_pts','mean_pm'])

df_compare.sort_values('mean_pm')
df_compare[df_compare['opp_player']=='Boban Marjanovic']



df_compare.groupby('opp_player').mean().sort_values('mean_pm')
```
